1. Faça um programa que mostra os numeros entre 1 e 500.

2. Faça um programa que mostra os números do 2 até o 1000, porém apenas os números pares.

3. Faça um programa que inicie uma contagem regressiva a partir de um número especificado pelo usuário até zero, utilizando um loop while.

4. Faça um programa que peça para o usuário o login e a senha, o laço irá repetir enquanto o login e a senha forem incorretos. O usuário terá 3 tentativas apenas.

5. Faça um programa que peça um número de 1 a 10 e exibe a tabuada. Por exemplo:</br>

5 x 1 = 5</br>
5 x 2 = 10</br>
5 x 3 = 15</br>
...</br>
5 x 10 = 50

```python
for i in range(1, 501):
    print(i)
```

```python
for i in range(2, 1001, 2):
    print(i)
```

```python
num = int(input("Digite um número para começar a contagem regressiva: "))

while num >= 0:
    print(num)
    num -= 1
```

```python
numero = int(input("Digite um número de 1 a 10: "))

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

```python
tentativas = 3
login_correto = "usuario"
senha_correta = "senha123"

while tentativas > 0:
    login = input("Digite o login: ")
    senha = input("Digite a senha: ")

    if login == login_correto and senha == senha_correta:
        print("Login bem sucedido!")
        break
    else:
        print("Login ou senha incorretos.")
        tentativas -= 1

if tentativas == 0:
    print("Número máximo de tentativas excedido.")
```
