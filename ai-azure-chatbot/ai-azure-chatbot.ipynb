# Usar o modelo de Respostas s Perguntas com o Language Studio

Nesse exerccio, o objetivo  criar e treinar uma base de dados de perguntas e respostas usando o Language Studio para um bot de atendimento ao cliente da Margie's Travel, uma agncia de viagens fictcia. A base de dados ser feita a partir da pgina de perguntas frequentes do site da empresa. </br></br>

A primeira etapa  criar uma base de dados de conhecimento com pares de perguntas e respostas, que o bot usar para interpretar e responder perguntas dos usurios. O Language Studio do Azure oferece recursos de resposta a perguntas, permitindo criar essa base de dados inserindo informaes manualmente ou importando de documentos ou pginas da web existentes.

# Criar um recurso de Linguagem

Para usar a resposta s perguntas, voc precisa de um recurso de**Linguagem**.

1. Em outra guia do navegador, abra o portal do Azure em[https://portal.azure.com](https://portal.azure.com/?azure-portal=true), entrando com a conta Microsoft associada  sua assinatura do Azure.
    
2. Clique no boto**Criar um recurso**e pesquise_Servio de linguagem_. Selecione**criar**um plano**Servio de linguagem**. Voc ser levado para uma pgina para**Selecionar recursos adicionais**. Use as configuraes a seguir:
    - **Selecione os Recursos Adicionais**:
        - **Recursos padro**:_mantenha os recursos padro_.
        - **Recursos personalizados**:_selecione a opo respostas s perguntas personalizadas_.
    - Selecione**Continuar para criar seu recurso**

image.png

3. Na pgina**Criar linguagem**, especifique as seguintes configuraes:
    - **Detalhes do projeto**
        - **Assinatura**:_sua assinatura do Azure_.
        - **Grupo de recursos**:_selecione um grupo de recursos existente ou crie um novo_.
    - **Detalhes da Instncia**
        - **Regio**:_Selecionar uma regio_
        - **Nome**:_um nome exclusivo para o recurso de Linguagem_.
        - **Tipo de preo**: S (mil chamadas por minuto)
    - **Respostas s perguntas personalizadas**
        - **Localizao do Azure Search**:_qualquer localizao disponvel_.
        - **Nvel de preo do Azure Search**: F Gratuito (3 ndices)  (_Se esse nvel no estiver disponvel, selecione Bsico_)
    - **Aviso de IA Responsvel**
        - **Ao marcar esta caixa, declaro que analisei e confirmo os termos do Aviso de IA responsvel**:_selecionada_.
4. Selecione**Examinar e Criar**e, em seguida, selecione**Criar**. Aguarde a implantao do servio de Linguagem que dar suporte  sua base de dados de conhecimento de respostas s perguntas personalizadas.


> **Observao**: caso j tenha provisionado um recurso do**Azure Cognitive Search**de camada gratuita, talvez a cota no permita a criao de outro. Nesse caso, selecione uma camada diferente de**Free F**.

# Crie um novo projeto

1. Em uma nova guia do navegador, abra o portal do Language Studio em[https://language.azure.com](https://language.azure.com/?azure-portal=true)e entre usando a conta Microsoft associada  sua assinatura do Azure.
2. Se for solicitado a escolher um recurso de idioma, selecione as seguintes configuraes:
    
    - **Diretrio do Azure**:_O diretrio do Azure que contm a sua assinatura_.
    - **Assinatura do Azure**:_sua assinatura do Azure_.
    - **Recurso de Linguagem**:_O recurso de Linguagem que voc criou anteriormente._
    
    Se voc**_no_**foi solicitado a escolher um recurso de idioma, pode ser porque voc tem vrios recursos de idioma em sua assinatura; nesse caso:
    
    1. Na barra na parte superior da pgina, selecione**Configuraes ()**.
    2. Na pgina**Configuraes**, exiba a guia**Recursos**.
    3. Selecione o recurso de idioma que voc acabou de criar e clique em**Alternar recurso**.
    4. Na parte superior da pgina, clique em**Language Studio**para retornar  pgina inicial do Language Studio.
3. Na parte superior do portal do Language Studio, no menu**Criar**, selecione**Respostas s perguntas personalizadas**.

image.png

image.png

4. Na pgina**Escolher configurao de idioma para o_recurso_**, selecione**Quero selecionar o idioma quando criar um projeto neste recurso**e clique em**Avanar**.

image.png

5. Na pgina**Inserir informaes bsicas**, insira os seguintes detalhes e clique em**Avanar**:
    - **Recurso de idioma**:_escolha o seu recurso de idioma_.
    - **Recurso de pesquisa do Azure**:_escolha o recurso de pesquisa do Azure_.
    - **Nome**:`MargiesTravel`
    - **Descrio**:`A simple knowledge base`
    - **Idioma de origem**: ingls
    - **Resposta padro quando nenhuma resposta  retornada**:`No answer found`
6. Na pgina**Revisar e concluir**, selecione**Criar projeto**.


7. Voc ser levado  pgina**Gerenciar origens**. Clique em**Adicionar origem**e selecione**URLs**.

image.png


8. **Na caixa Adicionar URLs**, selecione**+ Adicionar URL**. Digite o seguinte e selecione**Adicionar tudo**:
    - **Nome da URL**:`MargiesKB`
    - **URL**:`https://raw.githubusercontent.com/MicrosoftLearning/mslearn-ai-fundamentals/main/data/natural-language/margies_faq.docx`
    - **Classificar estrutura de arquivo**:_Detectar automaticamente_
9. Selecione**Adicionar tudo.**

# Editar a base de dados de conhecimento

Sua base de dados de conhecimento se baseia nos detalhes contidos no documento de perguntas frequentes e em algumas respostas pr-definidas. Voc pode adicionar pares personalizados de perguntas e respostas para complement-los.

1. Expanda o painel esquerdo e selecione**Editar base de dados de conhecimento**. Em seguida, selecione**+**para adicionar um novo par de perguntas.


image.png

2. Na caixa de dilogo**Adicionar um novo par de respostas e perguntas**, no**tipo de Pergunta** `Hello`e no **tipo de Resposta** `Hi`, selecione **Concludo**.


image.png

3. Expanda**Perguntas alternativas**e selecione**+ Adicionar pergunta alternativa**. Em seguida, insira`Hiya`como uma frase alternativa para Ol.


image.png

4. Na parte superior do painel**Pares de perguntas e respostas**, selecione**Salvar**para salvar sua base de dados de conhecimento.

# Treinar e testar a base de dados de conhecimento

Agora que voc tem uma base de dados de conhecimento, pode test-la.

1. Na parte superior do painel**Pares de perguntas e respostas**, selecione**Testar**para testar sua base de dados de conhecimento.
2. No painel de teste, na parte inferior, insira a mensagem`Hi`. A resposta_Oi_deve ser retornada.
3. No painel de teste, na parte inferior, insira a mensagem`I want to book a flight`. Uma resposta apropriada das perguntas frequentes deve retornar.

image.png

> **Observao**A resposta inclui uma_resposta curta_, bem como uma_passagem de resposta_mais detalhada. A passagem de resposta mostra o texto completo no documento de perguntas frequentes da pergunta com a correspondncia mais prxima, enquanto a resposta curta  extrada da passagem de maneira inteligente. Voc pode controlar se quer mostrar a resposta curta usando a caixa de seleo**Exibir resposta curta**na parte superior do painel de teste.

4. Tente outra pergunta, como`How can I cancel a reservation?`
5. Quando terminar de testar a base de dados de conhecimento, clique em**Testar**para fechar o painel de teste.

# Criar um bot para a base de dados de conhecimento

A base de dados de conhecimento fornece um servio de back-end que os aplicativos cliente podem usar para responder a perguntas por meio de algum tipo de interface do usurio. Normalmente, esses aplicativos cliente so bots. Para disponibilizar a base de dados de conhecimento a um bot, voc deve public-la como um servio que possa ser acessado por HTTP. Ento, voc pode usar o Servio de Bot do Azure para criar e hospedar um bot que usa a base de dados de conhecimento para responder a perguntas do usurio.

1. No painel esquerdo, selecione**Implantar base de dados de conhecimento**.
2. Na parte superior da pgina, selecione**Implantar**. Uma caixa de dilogo perguntar se voc deseja implantar o projeto. Selecione**Implantar**.

image.png

1. Aps a implantao do servio, clique em**Criar um bot**. Isso abre o portal do Azure em uma nova guia do navegador para que voc possa criar um Bot de Aplicativo Web em sua assinatura do Azure.


image.png

2. No portal do Azure, crie um**Bot de Aplicativo Web**. (Voc pode ver uma mensagem de aviso para marcar que a origem do modelo  confivel. Voc no precisa tomar nenhuma ao para essa mensagem.) Continue atualizando as seguintes configuraes:
    
    - **Detalhes do projeto**
        - **Assinatura**:_sua assinatura do Azure_
        - **Grupo de recursos**:_o grupo de recursos que contm seu recurso de Linguagem_
    - **Detalhes da instncia**
        - **Localizao do grupo de recursos**:_o mesmo local do servio de linguagem_.
    - **Bot do Azure**
        - **Identificador do bot**:_insira um nome exclusivo para o bot_(_preenchido previamente_)
    - **Escolha o tipo de preo**
        - **Tipo de preo**: Gratuito (F0) (Talvez seja necessrio selecionar_Alterar plano_)
    - **ID do Aplicativo da Microsoft**
        - **Tipo de criao**:_Selecione **Criar uma nova identidade gerenciada atribuda pelo usurio**
3. Selecione**Avanar**para continuar atualizando as configuraes.
    - **Servio de Aplicativo**
        - **Nome do aplicativo**:_o mesmo que a**Ala do bot**com**.azurewebsites.net**anexado automaticamente_
        - **Linguagem do SDK**:_escolha C# ou Node.js_
    - **Plano do Servio de Aplicativo**
        - **Tipo de Criao**:_Selecione**Criar novo plano de servio de aplicativo**_
    - **Configuraes do aplicativo**
        - **Chave de Recurso de Linguagem**:_voc precisar copiar sua chave de recurso de linguagem e col-la aqui_:
            - Em outra guia do navegador, acesse o portal do Azure em[https://portal.azure.com](https://portal.azure.com/?azure-portal=true).
            - Navegue at o recurso de servio de linguagem.
            - Na pgina**Chaves e Ponto de Extremidade**, copie uma das chaves
            - Cole-a aqui.
        - **Nome do projeto de linguagem**: MargiesTravel
        - **Nome do host do ponto de extremidade do servio de linguagem**:_preenchido previamente com o ponto de extremidade do servio de idioma_
    - **Detalhes do servio de linguagem**
        - **ID da assinatura**:_preenchido previamente com sua ID de assinatura_
        - **Nome do Grupo de Recursos**:_preenchido previamente com o nome do grupo de recursos_
        - **Nome da conta**:_preenchido previamente com o nome do recurso_
4. Selecione**Criar**. Aguarde enquanto o bot  criado (o cone de notificao do canto superior direito, parecido com um sino, ficar animado enquanto voc espera). Em seguida, na notificao que diz que a implantao foi concluda, clique em**Ir para o recurso**(ou, na pgina inicial, selecione**Grupos de recursos**, abra o grupo de recursos em que voc criou o bot e selecione o recurso**Bot do Azure**.)



image.png

5. No painel  esquerda do bot, procure**Configuraes**, clique em**Testar no Webchat**e aguarde at que o bot exiba a mensagem**Ol e bem-vindo(a)**(a inicializao pode demorar alguns segundos).
6. Use a interface de chat de teste para garantir que o bot responda s perguntas da sua base de dados de conhecimento conforme o esperado. Por exemplo, tente enviar`I need to cancel my hotel`.
